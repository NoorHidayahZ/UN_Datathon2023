```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
```

```python
df = pd.read_excel('../References/Jadual-2-3-sungai-tercemar-2020.xlsx')
df.head()
```

```python
geolocator = Nominatim(user_agent="river_geocoder")
```

```python
# insert space in camel case
def insert_spaces_in_camel_case(input_string):
    output_string = input_string[0]  # Initialize the output string with the first character

    for i in range(1, len(input_string)):
        if input_string[i].isupper() and not input_string[i - 1].isspace():
            output_string += ' '  # Add a space before the uppercase letter
        output_string += input_string[i]

    return output_string

df['river_name'] = df['river'].apply(insert_spaces_in_camel_case)
```

```python
df['Coordinates'] = df['river_name'].apply(lambda x: geolocator.geocode(x) if x is not None else None)
mask = df['Coordinates'].isnull()
df['Coordinates'][mask] = df['river-basin'][mask].apply(lambda x: geolocator.geocode(insert_spaces_in_camel_case(x)) if x is not None else None)
```

```python
# Correction for row 0 and 2
df.loc[0,'Coordinates'] = None
df.loc[2,'Coordinates'] = None
```

```python
# Extract latitude and longitude from the geocoded results
df['Latitude'] = df['Coordinates'].apply(lambda x: x.latitude if x is not None else None)
df['Longitude'] = df['Coordinates'].apply(lambda x: x.longitude if x is not None else None)

# df.loc[22,'Coordinates'] = a
df.loc[22,'Latitude'] = geolocator.geocode('Pasir Gudang').latitude
df.loc[22,'Longitude'] = geolocator.geocode('Pasir Gudang').longitude
df.loc[0,'Longitude'] = geolocator.geocode('Sg. Jawi').longitude
df.loc[0,'Latitude'] = geolocator.geocode('Sg. Jawi').latitude
df.loc[2,'Longitude'] = geolocator.geocode('Sg. Jawi').longitude
df.loc[2,'Latitude'] = geolocator.geocode('Sg. Jawi').latitude
df.loc[26,'Longitude'] = geolocator.geocode('Sg. Tebrau').longitude
df.loc[26,'Latitude'] = geolocator.geocode('Sg. Tebrau').latitude
# Drop rows with missing coordinates
# data = data.dropna(subset=['Latitude', 'Longitude'])
df.to_excel('../References/Jadual-2-3-sungai-tercemar-2020-geocoded.xlsx')
df
```
